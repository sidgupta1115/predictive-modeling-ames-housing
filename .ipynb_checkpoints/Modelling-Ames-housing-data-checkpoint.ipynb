Quantify various Qual/Condition values

> **Note:** I did this project before I learned one-hotencoding for categorical variables

```python
quantified_features= {"BsmtCond" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "BsmtExposure" : {"No" : 0, "Mn" : 1, "Av": 2, "Gd" : 3},
                       "BsmtFinType1" : {"No" : 0, "Unf" : 1, "LwQ": 2, "Rec" : 3, "BLQ" : 4, 
                                         "ALQ" : 5, "GLQ" : 6},
                       "BsmtFinType2" : {"No" : 0, "Unf" : 1, "LwQ": 2, "Rec" : 3, "BLQ" : 4, 
                                         "ALQ" : 5, "GLQ" : 6},
                       "BsmtQual" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA": 3, "Gd" : 4, "Ex" : 5},
                       "ExterCond" : {"Po" : 1, "Fa" : 2, "TA": 3, "Gd": 4, "Ex" : 5},
                       "ExterQual" : {"Po" : 1, "Fa" : 2, "TA": 3, "Gd": 4, "Ex" : 5},
                       "FireplaceQu" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "Functional" : {"Sal" : 1, "Sev" : 2, "Maj2" : 3, "Maj1" : 4, "Mod": 5, 
                                       "Min2" : 6, "Min1" : 7, "Typ" : 8},
                       "GarageCond" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "GarageQual" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "HeatingQC" : {"Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "KitchenQual" : {"Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "LandSlope" : {"Sev" : 1, "Mod" : 2, "Gtl" : 3},
                       "LotShape" : {"IR3" : 1, "IR2" : 2, "IR1" : 3, "Reg" : 4},
                       "PavedDrive" : {"N" : 0, "P" : 1, "Y" : 2},
                       "Street" : {"Grvl" : 1, "Pave" : 2},
                       "Utilities" : {"ELO" : 1, "NoSeWa" : 2, "NoSewr" : 3, "AllPub" : 4}}

house = house.replace(quantified_features)
```

```python
#split into numerical and catergorical
categorical_features = house.select_dtypes(include = ["object"])
numerical_features = house.select_dtypes(exclude = ["object"])
```

```python
# Fill in missing values
categorical_features = categorical_features.fillna("No")
numerical_features = numerical_features.fillna(numerical_features.median())
```

**Feature Engineering:**

```python
# 1) Composite Score by taking the product of Quality and Condition metrics

# Overall score of the house
numerical_features["OverallScore"] = numerical_features["OverallQual"] * numerical_features["OverallCond"]
# Overall score of the garage
numerical_features["GarageScore"] = numerical_features["GarageQual"] * numerical_features["GarageCond"]
# Overall score of the exterior
numerical_features["ExterScore"] = numerical_features["ExterQual"] * numerical_features["ExterCond"]
# Overall kitchen score
numerical_features["KitchenScore"] = numerical_features["KitchenAbvGr"] * numerical_features["KitchenQual"]
# Overall fireplace score
numerical_features["FireplaceScore"] = numerical_features["Fireplaces"] * numerical_features["FireplaceQu"]
# Overall garage score
numerical_features["GarageScore"] = numerical_features["GarageArea"] * numerical_features["GarageQual"]
# Overall Bsmt score
numerical_features["BsmtScore"] = numerical_features["BsmtQual"] * numerical_features["BsmtExposure"]*\
                                np.mean(numerical_features["BsmtFinType1"]+ numerical_features["BsmtFinType2"])

#2) Sum up total Bathrooms and Square Ft

# Total number of bathrooms
numerical_features["TotalBath"] = numerical_features["BsmtFullBath"] + (0.5 * numerical_features["BsmtHalfBath"]) + \
                                    numerical_features["FullBath"] + (0.5 * numerical_features["HalfBath"])
# Total SF for house (incl. basement)
numerical_features["InteriorSF"] = numerical_features["GrLivArea"] + numerical_features["TotalBsmtSF"]
# Total SF for porch
numerical_features["ExteriorSF"] = numerical_features["OpenPorchSF"] + numerical_features["EnclosedPorch"] + \
                                    numerical_features["3SsnPorch"] + numerical_features["ScreenPorch"]
```

```python
# Concatinate them back together
house=pd.concat([numerical_features, categorical_features], axis=1)
```

**Building A Model**

First I wanna train our model only on the static features from the dataframe and remove the renovatable ones

```python
house=house.drop([col for col in house.columns if ('Qual' in col)], axis=1)
house=house.drop([col for col in house.columns if ('QC' in col)], axis=1)
house=house.drop([col for col in house.columns if ('Cond' in col)], axis=1)                 
house=house.drop([col for col in house.columns if ('Qu' in col)], axis=1)
house=house.drop([col for col in house.columns if ('Quality' in col)], axis=1)
house=house.drop([col for col in house.columns if ('Score' in col)], axis=1)
house=house.drop([col for col in house.columns if ('BsmtFin' in col)], axis=1)
house=house.drop([col for col in house.columns if ('BsmtUnf' in col)], axis=1)
house=house.drop([col for col in house.columns if ('BsmtFin' in col)], axis=1)

```

```python
house.shape #our model will be trained on 59 features
```

```python
#Dummy categorical variables before we use a regression model
categorical_features = house.select_dtypes(include = ["object"])
categorical_features=pd.get_dummies(categorical_features)
numerical_features = house.select_dtypes(exclude = ["object"])

house=pd.concat([numerical_features, categorical_features], axis=1)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV, ElasticNetCV, Lasso, ElasticNet
from sklearn.feature_selection import SelectFromModel, SelectKBest, f_regression
```

```python
#We'll be training our data on houses sold before 2010 and testing it on houses sold after 2010

train=house[house['YrSold']<2010]
test=house[house['YrSold'] ==2010]

X_train=train.drop(['SalePrice'], axis=1)
X_test=test.drop(['SalePrice'], axis=1)
y_train=train['SalePrice']
y_test=test['SalePrice']

X_train.shape, X_test.shape, y_train.shape, y_test.shape # Make sure our dimensions match
```

I used 2 linear regression models Lasso and ElasticNet I chose  Lasso as my first regularization method so I could eliminate features that dont have much of an affect I used the ElasticNet after to see how the effect of adding Ridge to the Lasso would effect the performace of my model

> **Note:** I did each model twice, once using a GridSearchCV because I wanted to practice making models both ways

```python
lasso = LassoCV(alphas = np.arange(.0001, 1, 10),  cv = 5)

lasso.fit(X_train, y_train)
alpha = lasso.alpha_
score = lasso.score(X_train, y_train)
testscore = lasso.score(X_test, y_test)
predictions = lasso.predict(X_test)

print("Best alpha :", alpha)
print("Best train score :", score)
print("Best test score :", testscore)
```

```python
lasso_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('regr', Lasso())
])


lasso_params = {
    'regr__alpha':np.arange(.0001, 1, 10)
}

lasso_gs = GridSearchCV(lasso_pipe,
                      param_grid=lasso_params,
                      n_jobs=-1,
                      cv=ShuffleSplit(5)
                     )
lasso_gs.fit(X_train, y_train)
```

```python
lasso_gs.best_params_
```

```python
lasso_gs.best_score_
```

```python
pd.DataFrame(lasso_gs.cv_results_).T
```

```python
elastic_net= ElasticNetCV(l1_ratio = np.arange(.1, 1, 10),
                          alphas = np.arange(.0001, 1, 10) , cv = 5)
elastic_net.fit(X_train, y_train)
alpha = elastic_net.alpha_
ratio = elastic_net.l1_ratio_
score = lasso.score(X_train, y_train)
testscore = elastic_net.score(X_test, y_test)
predictions = elastic_net.predict(X_test)
print("Best l1_ratio :", ratio)
print("Best alpha :", alpha )
print("Best train score :", score)
print("Best test score :", testscore)
```

```python
elasticnet_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('regr', ElasticNet())
])


elasticnet_params = {
    'regr__l1_ratio': np.arange(.1, 1, 10),
    'regr__alpha':np.arange(.0001, 1, 10)
}

elasticnet_gs = GridSearchCV(elasticnet_pipe,
                      param_grid=elasticnet_params,
                      n_jobs=-1,
                      cv=ShuffleSplit(5)
                     )
elasticnet_gs.fit(X_train, y_train)
```

```python
elasticnet_gs.best_params_
```

```python
elasticnet_gs.best_score_
```

Both our Lasso and Elastic Net models are decent predictive models based on their test scores however the Lasso tends to perform slightly better and may be preferred 

## 2. Determine any value of *changeable* property characteristics unexplained by the *fixed* ones.

---

Now that I have a model that estimates the price of a house based on its fixed characteristics, we can move forward to determine what are the costs/benefits of quality, condition, and renovations

**Goal:**
Evaluate the effect in dollars of a model based on renovate-able features. 

```python
#Lets Reload our data, clean it and engineer the features
```

```python
house = pd.read_csv('./housing.csv', index_col=0)

house = house.drop(['PoolQC','MiscFeature','Alley','Fence'], axis=1)

house = house[(house['MSZoning'] == "RH") | (house['MSZoning'] == "RL") |\
              (house['MSZoning'] == "RP") | (house['MSZoning'] == "RM")]

house = house.replace({"BsmtCond" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "BsmtExposure" : {"No" : 0, "Mn" : 1, "Av": 2, "Gd" : 3},
                       "BsmtFinType1" : {"No" : 0, "Unf" : 1, "LwQ": 2, "Rec" : 3, "BLQ" : 4, 
                                         "ALQ" : 5, "GLQ" : 6},
                       "BsmtFinType2" : {"No" : 0, "Unf" : 1, "LwQ": 2, "Rec" : 3, "BLQ" : 4, 
                                         "ALQ" : 5, "GLQ" : 6},
                       "BsmtQual" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA": 3, "Gd" : 4, "Ex" : 5},
                       "ExterCond" : {"Po" : 1, "Fa" : 2, "TA": 3, "Gd": 4, "Ex" : 5},
                       "ExterQual" : {"Po" : 1, "Fa" : 2, "TA": 3, "Gd": 4, "Ex" : 5},
                       "FireplaceQu" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "Functional" : {"Sal" : 1, "Sev" : 2, "Maj2" : 3, "Maj1" : 4, "Mod": 5, 
                                       "Min2" : 6, "Min1" : 7, "Typ" : 8},
                       "GarageCond" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "GarageQual" : {"No" : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "HeatingQC" : {"Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "KitchenQual" : {"Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5},
                       "LandSlope" : {"Sev" : 1, "Mod" : 2, "Gtl" : 3},
                       "LotShape" : {"IR3" : 1, "IR2" : 2, "IR1" : 3, "Reg" : 4},
                       "PavedDrive" : {"N" : 0, "P" : 1, "Y" : 2},
                       "Street" : {"Grvl" : 1, "Pave" : 2},
                       "Utilities" : {"ELO" : 1, "NoSeWa" : 2, "NoSewr" : 3, "AllPub" : 4}}
                     )


categorical_features = house.select_dtypes(include = ["object"])
categorical_features = categorical_features.fillna("No")
categorical_features=pd.get_dummies(categorical_features)
numerical_features = house.select_dtypes(exclude = ["object"])
numerical_features = numerical_features.fillna(numerical_features.median())

# 2* Combinations of existing features
# Overall quality of the house
numerical_features["OverallScore"] = numerical_features["OverallQual"] * numerical_features["OverallCond"]
# Overall quality of the garage
numerical_features["GarageScore"] = numerical_features["GarageQual"] * numerical_features["GarageCond"]
# Overall quality of the exterior
numerical_features["ExterScore"] = numerical_features["ExterQual"] * numerical_features["ExterCond"]
# Overall kitchen score
numerical_features["KitchenScore"] = numerical_features["KitchenAbvGr"] * numerical_features["KitchenQual"]
# Overall fireplace score
numerical_features["FireplaceScore"] = numerical_features["Fireplaces"] * numerical_features["FireplaceQu"]
# Overall garage score
numerical_features["GarageScore"] = numerical_features["GarageArea"] * numerical_features["GarageQual"]
# Total number of bathrooms
numerical_features["TotalBath"] = numerical_features["BsmtFullBath"] + (0.5 * numerical_features["BsmtHalfBath"]) + \
numerical_features["FullBath"] + (0.5 * numerical_features["HalfBath"])
# Total SF for house (incl. basement)
numerical_features["InteriorSF"] = numerical_features["GrLivArea"] + numerical_features["TotalBsmtSF"]
# Total SF for porch
numerical_features["ExteriorSF"] = numerical_features["OpenPorchSF"] + numerical_features["EnclosedPorch"] + \
numerical_features["3SsnPorch"] + numerical_features["ScreenPorch"]

house=pd.concat([numerical_features, categorical_features], axis=1)
house["SalePrice"] = np.log(house["SalePrice"])
```

```python
# Lets make a dataframe only with the renovatable feature for Quality and Condition
Qual=house.loc[:, house.columns.str.contains('Qual')]
QC=house.loc[:, house.columns.str.contains('QC')]
Cond=house.loc[:, house.columns.str.contains('Cond')]
Qu=house.loc[:, house.columns.str.contains('Qu')]
Quality=house.loc[:, house.columns.str.contains('Quality')]
BsmtFin=house.loc[:, house.columns.str.contains('BsmtFin')]
BsmtUnf=house.loc[:, house.columns.str.contains('BsmtUnf')]
Score=house.loc[:, house.columns.str.contains('Score')]
YrSold=house.loc[:, house.columns.str.contains('YrSold')]
SalePrice=house.loc[:, house.columns.str.contains('SalePrice')]

house=pd.concat([Qual, QC, Cond, Qu, Quality, BsmtFin, BsmtUnf, YrSold, SalePrice], axis=1)
house=house.T.drop_duplicates().T
```

I'm going to make 2 subsets of data to test on:
1. Poorer quality homes - defined as homes with an OverallQual in the lower 25th percentile or lower
2. Better quality - homes with an OverallQual greater that the mean

```python
X_test_poor=house[house["OverallQual"] < np.percentile(house["OverallQual"], 25)]
X_test_better=house["OverallQual"]>house["OverallQual"].mean()
```

```python
#Lets use a Lasso model on the renovatable features
```

```python
train=house[house['YrSold']<2010]
test=house[house['YrSold'] ==2010]

X_train=train.drop(['SalePrice'], axis=1)
X_test=test.drop(['SalePrice'], axis=1)
y_train=train['SalePrice']
y_test=test['SalePrice']
```

Lets compare predictations of SalePrice for OverallQual that are less than the 25th percentile (Poor Condtion) for 
Qual/Cond/Score to SalePrice predications that perform better than the median Overall Qual(Better Condition)

```python

X_test_poor=house[house["OverallQual"] <= np.percentile(house["OverallQual"], 25)].drop(['SalePrice'], axis=1)
X_test_better=house[house["OverallQual"]>house["OverallQual"].mean()].drop(['SalePrice'], axis=1)
X_train.shape, X_test_poor.shape, X_test_better.shape, y_train.shape, y_test.shape
```

We'll plot these predications after. If the SalePrice values for the better conditioned houses are substantially better than the houses in poorer condtion we can recommend investing in renovations.

```python
lasso = LassoCV(alphas = np.arange(.0001, 1, 10),  cv = 5)

lasso.fit(X_train, y_train)
alpha = lasso.alpha_
score = lasso.score(X_train, y_train)
testscore = lasso.score(X_test, y_test)

predictions_poor = np.exp(lasso.predict(X_test_poor)) #exponentiate it to get values back in dollar values
predictions_better = np.exp(lasso.predict(X_test_better))

print("Best alpha :", alpha)
print("Best train score :", score)
print("Best test score :", testscore)
```

```python
plt.hist([predictions_poor, predictions_better], histtype='bar', 
        align='mid', label=['Poorer Quality','Better Quality']) #distribution
plt.title("Distrubution of predictions")
plt.legend(loc = "upper right")
plt.show()
```

It looks like better quality houses definitely sell for more than poorer quality houses

Let's take a look at our median predictions

```python
np.median(predictions_poor)
```

```python
np.median(predictions_better)
```

```python
#Check percent increase
(np.median(predictions_better)-np.median(predictions_poor))/np.median(predictions_poor)
```

The houses in the poorer 25th percentile about can increase by 82% if you rennovate it until it is in avg to excellent quality.


Said another way: you can put up to 82% of the price of a poor quality house in rennovations to get it in better condition, and still expect break even. Therefore even minimal rennovations can yield a higher SalePrice and definitiley warrant making investments in these homes.
